The df.describe() methods provides the 'count' 'mean' 'median' 'std' 'min' '25%' '50%' '75%' and 'max'

Definitions 

'count' - the number of not empty values or non missing values in a dataset
'mean'  - the average(mean) value 
'median'- the middle value of a sorted list of numbers
'std'   - the standard deviation
'min'   - the minimum value 
'25%'   - the 25th percintile (the value which below 25% of the data falls - less than or equal to) also known as quartile 1 (Q1)
'50%'   - the 50th percintile (the value which below 50% of the data falls - less than or equal to)
'75%'   - the 75th percintile (the value which below 75% of the data falls - less then or equal to) (quartile 3)
'max'   - the maximum value

In this scenario count lets us know how many passengers are in the survived columns - (how many male and females survival status are recorded) Mean then tells us the average number of passengers that survived in each group the higher the average/mean the more people that survived.
